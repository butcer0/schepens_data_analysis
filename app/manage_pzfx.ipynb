```python
xml_as_bytes = sample_xml('rb')
print('[etree lxml] Starting to parse XML')

tree = etree_lxml.fromstring(xml_as_bytes)
xml_etree_lxml = tree.findall('./TableSequence', {})
```

```python
def write_row_title_columns(row_titles):
    sub_column_root = etree_lxml.Element('Subcolumn')
    for row_title in row_titles:
        row = etree_lxml.SubElement(sub_column_root, 'd')
        row.text = row_title

    return sub_column_root

print(etree_lxml.tostring(write_row_title_columns(['0', '1', '2+'])))
```

```python
def write_subcolumn(root, data):
    # root = etree_lxml.Element('Subcolumn')
    for row_data in data:
        row = etree_lxml.SubElement(root, 'd')
        row.text = row_data

    return root

print(etree_lxml.tostring(write_row_title_columns(['0', '1', '2+'])))
```

```python
def write_y_column(title, attributes, data):
    y_column_root = etree_lxml.Element('YColumn')
    for attribute_key, attribute_value in attributes.items():
        y_column_root.set(attribute_key, attribute_value)

    sub_column_root = etree_lxml.SubElement(y_column_root, 'Subcolumn')
    for row_data in data:
        row = etree_lxml.SubElement(sub_column_root, 'd')
        row.text = row_data

    return y_column_root

print(etree_lxml.tostring(write_y_column('Con1', {'Width':'588','Decimals':'6','Subcolumns':'3'}, ['Count', 'Total Volume', 'Mean Volume'])))

```
